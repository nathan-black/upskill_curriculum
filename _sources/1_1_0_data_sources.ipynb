# Data Sources
New data is constantly being generated by the things you do and the processes that surround your actions. Recognizing your digital footprint and the data trails created around you will open your eyes to how you are viewed by others through the lens of data. In this section, you’ll learn to how to use data to describe people, objects, places, and actions. You’ll learn to identify data observations, the attributes that describe them, and how data values vary across observations.
